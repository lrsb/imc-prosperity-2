```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, probplot
import pandas as pd
```

```python
# Extract prices of AMETHYSTS
df = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_-1.csv', sep=';')
amethysts_prices = df.loc[ df['product'] == 'AMETHYSTS', 'bid_price_1'].to_list()

df = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_-2.csv', sep=';')
amethysts_prices += df.loc[ df['product'] == 'AMETHYSTS', 'bid_price_1'].to_list()

df = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_0.csv', sep=';')
amethysts_prices += df.loc[ df['product'] == 'AMETHYSTS', 'bid_price_1'].to_list()

# Plot histogram
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(amethysts_prices, bins=10, edgecolor='black')
plt.title('Histogram of Prices for AMETHYSTS')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Plot Q-Q plot
plt.subplot(1, 2, 2)
probplot(amethysts_prices, dist="norm", plot=plt)
plt.title('Q-Q Plot of Prices for AMETHYSTS')

plt.tight_layout()
plt.show()

# Perform Shapiro-Wilk test
stat, p = shapiro(amethysts_prices)
print(f"Shapiro-Wilk Test Statistic: {stat:.3f}, p-value: {p:.5f}")
```
