```python
lijst = ['Pizza', 'Wasabi', 'Snow', 'Shell']
[first + '>' + second for first in lijst for second in lijst]
```

```python
conversions = {
    'Pizza>Pizza': 1,
    'Pizza>Wasabi': 0.48,
    'Pizza>Snow': 1.52,
    'Pizza>Shell': 0.71,
    'Wasabi>Pizza': 2.05,
    'Wasabi>Wasabi': 1,
    'Wasabi>Snow': 3.26,
    'Wasabi>Shell': 1.56,
    'Snow>Pizza': 0.64,
    'Snow>Wasabi': 0.3,
    'Snow>Snow': 1,
    'Snow>Shell': 0.46,
    'Shell>Pizza': 1.41,
    'Shell>Wasabi': 0.61,
    'Shell>Snow': 2.08,
    'Shell>Shell': 1
}

for first in conversions:
    if 'Shell>' in first:
        score_first = conversions[first]
        print('1 -', score_first, first)
        
        for second in conversions:
            if first.split('>')[1] + '>' in second:
                score_second = conversions[second] * score_first
                print('2 -', score_second, second)

                for third in conversions:
                    if second.split('>')[1] + '>' in third:
                        score_third = conversions[third] * score_second
                        print('3 -', score_third, third)
                
                        for fourth in conversions:
                            if third.split('>')[1] + '>' in fourth:
                                score_fourth = conversions[fourth] * score_third
                                print('4 -', score_fourth, fourth)

                                for fifth in conversions:
                                    if fourth.split('>')[1] + '>' + 'Shell' == fifth:
                                        score_fifth = conversions[fifth] * score_fourth
                                        print('5 -', score_fifth, fifth)
```
