```python
import numpy as np
import pandas as pd

pd.options.plotting.backend = 'plotly'

prices = []
trades = []

last_timestamp = -100
for day in range(1, 4):
    p = pd.read_csv(f'../../data/prices_round_4_day_{day}.csv', sep=';')
    p['timestamp'] = p['timestamp'] + last_timestamp + 100
    prices.append(p)

    t = pd.read_csv(f'../../data/trades_round_4_day_{day}_nn.csv', sep=';')
    t['timestamp'] = t['timestamp'] + last_timestamp + 100
    trades.append(t)

    last_timestamp = np.max(p['timestamp'])

prices = pd.concat(prices, axis=0, ignore_index=True)
trades = pd.concat(trades, axis=0, ignore_index=True)

prices
```

```python
prices[prices['product'] == 'COCONUT'].plot(y=['mid_price'])
```

```python
prices[prices['product'] == 'COCONUT_COUPON'].plot(y=['mid_price'])
```

```python
import statsmodels.api as sm

X = prices[prices['product'] == 'COCONUT_COUPON']['mid_price'].values
y = prices[prices['product'] == 'COCONUT']['mid_price'].values
X = sm.add_constant(X)

model = sm.OLS(y, X, missing='drop').fit()
model.predict(X)

print(model.summary())
print(model.params)
```
